```python
import os
import torch
import pandas as pd
import scanpy as sc
from SPHERE.preprocess import *
from SPHERE.SPHERE import *
from SPHERE.utils import *
import numpy as np
device = torch.device('cuda:3' if torch.cuda.is_available() else 'cpu')
```

```python
adata = sc.read('./Cortex.h5ad')
adata_refs = [
    adata[adata.obs['batch'] == b].copy()
    for b in adata.obs['batch'].unique()
]
adata = [create_adata_from_latent(ad) for ad in adata_refs]
normalize_spatial_list(adata)
```

```python
model = SPHERE(adata, device=device, alignment=True, rwrIter = 100, rwIter = 100, inIter = 50, outIter = 100, alpha = 0.5, beta = 0.15,
        gamma = 0.7, l1 = 0.0005, l2 = 0.01, l3 = 0.0005, l4 = 0.00002)
```

```python
adata = model.train_align()
```

```python
adata_all = ad.concat(adata,label='batch')
```

```python
import matplotlib.pyplot as plt
import numpy as np

coords = adata_all.obsm['spatial']  
labels = adata_all.obs['annotation'].astype(str)
unique_labels = labels.unique()
layer_colors = {
    'ITL23GL': '#4CAF50',  
    'ITL4GL': '#3F7FBF',   
    'ITL5GL': '#8E44AD',   
    'ITL6GL': '#E67E22',   
    'L6bGL': '#E74C3C'     
}

label_to_color = {label: layer_colors.get(label, '#D3D3D3') for label in unique_labels}
cell_colors = labels.map(label_to_color)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    coords[:, 2], coords[:, 1], 1*coords[:, 0],  
    c=cell_colors,
    s=0.5,    
    alpha=0.8  
)

ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_zlabel('')

ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.grid(False)

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_facecolor('none')
fig.patch.set_facecolor('none')

for axis in ['x', 'y', 'z']:
    getattr(ax, f'{axis}axis').set_pane_color('none')

ax.view_init(elev=200, azim=310)

plt.tight_layout()
plt.show()
```
